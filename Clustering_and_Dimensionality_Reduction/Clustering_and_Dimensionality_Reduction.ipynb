# KMeans Clustering

```python
Image(url= "https://upload.wikimedia.org/wikipedia/commons/e/ea/K-means_convergence.gif", width=500, height=500)
```

```python
from sklearn import datasets
from sklearn.cluster import KMeans

# load in the iris dataset from the sklearn datasets
iris_df = datasets.load_iris()

print(iris_df)
```

```python
# instantiate the KMeans model
model = KMeans(n_clusters = 3)

# cluster the iris dataset
model.fit(iris_df.data)
```

Default parameters for KMeans in sklearn seen above. algorithm = 'auto'<br>
"auto” chooses “elkan” for dense data and “full” for sparse data.

```python
# Now we have clusters we can predict which cluster a new iris belongs to

predicted_label = model.predict([[7.2,3.5,0.8,1.6]])
```

```python
# or assign the entire dataset to each cluster

all_predictions = model.predict(iris_df.data)

print(predicted_label)
print(all_predictions)
```

# Hierarchical Clustering

```python
Image(url= "http://www.statisticshowto.com/wp-content/uploads/2016/11/clustergram.png", width=700, height=700)
```

```python
# Modules for hierarchical clustering
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import pandas as pd
```

```python
seeds_df = pd.read_csv(
    "https://raw.githubusercontent.com/vihar/unsupervised-learning-with-python/master/seeds-less-rows.csv")
    
# drop the grain species from the dataframe (target label?)
varieties = list(seeds_df.pop('grain_variety'))   #.pop returns the dataframe minus the specified column 
```

```python
# extract the seed measurements from the dataframe as a numpy array
samples = seeds_df.values
```

Hierarchical clustering is then carried out on the samples using the linkage() function with the method = 'complete'. Saving the results in a new variable mergings

```python
mergings = linkage(samples, method = 'complete')
```

Plot a dendrogram from the mergings labelled with their corresponding variety

```python
plt.rcParams["figure.figsize"] = (18,5)
dendrogram(mergings,
          labels = varieties,
          leaf_rotation = 90,
          leaf_font_size = 12)
plt.show()
```

# t-SNE (t-distributed Stochastic Neighbour Embedding)

```python
Image(url= "https://indico.io/blog/wp-content/uploads/2015/08/mnist.jpg", width=700, height=700)
```

```python
# get the t-SNE module from sklearn
from sklearn import datasets
from sklearn.manifold import TSNE

iris_df = datasets.load_iris()
```

```python
# TNSE has a learning rate parameter which should be optimised
model = TSNE(learning_rate = 100)
```

TSNE(n_components=2, perplexity=30.0, early_exaggeration=12.0, learning_rate=200.0, n_iter=1000, n_iter_without_progress=300, min_grad_norm=1e-07, metric=’euclidean’, init=’random’, verbose=0, random_state=None, method=’barnes_hut’, angle=0.5)

The learning rate for t-SNE is usually in the range [10.0, 1000.0]. If the learning rate is too high, the data may look like a ‘ball’ with any point approximately equidistant from its nearest neighbours. If the learning rate is too low, most points may look compressed in a dense cloud with few outliers. If the cost function gets stuck in a bad local minimum increasing the learning rate may help.

```python
# Perform TSNE on the iris dataset
transformed = model.fit_transform(iris_df.data)
```

```python
plt.rcParams["figure.figsize"] = (10,5)

x_axis = transformed[: , 0] # the first column
y_axis = transformed[:, 1] # the second column

# plot a scatter of the 2D t-SNE coloured by the iris target (species)
plt.scatter(x_axis, y_axis, c = iris_df.target)
```

# DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

```python
Image(url= "https://cdn-images-1.medium.com/max/2000/1*zbm_3K647rvNDmgL6HWUNQ.png", width=800, height=800)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA

from sklearn.datasets import load_iris

iris = load_iris()
```

```python
dbscan = DBSCAN()

dbscan.fit(iris.data)
```

```python
# transofrm the data with PCA
pca = PCA(n_components = 2).fit(iris.data)
pca_2d = pca.transform(iris.data)
```

```python
# plot the classes with different colours on a 2D scatter plot

for i in range(0, pca_2d.shape[0]):
    if dbscan.labels_[i] == 0:
        c1 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c = 'r', marker = '+')
    elif dbscan.labels_[i] == 1:
        c2 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='g', marker = 'o')
    elif dbscan.labels_[i] == -1:
        c3 = plt.scatter(pca_2d[i, 0], pca_2d[i,1], c = 'b', marker = '*')
        
plt.legend([c1, c2, c3], ['Cluster 1', 'Cluster 2', 'Noise'])
plt.title('DBSCAN finds 2 Clusters and Noise')
```

# PCA (Principal Component Analysis)

```python
Image(url= "http://www.nlpca.org/fig_pca_principal_component_analysis.png", width=800, height=800)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


from sklearn import decomposition
from sklearn import datasets

np.random.seed(5)

centers = [[1, 1], [-1, -1], [1, -1]]
iris = datasets.load_iris()
X = iris.data
y = iris.target

fig = plt.figure(1, figsize=(10, 7))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

plt.cla()
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)

for name, label in [('Setosa', 0), ('Versicolour', 1), ('Virginica', 2)]:
    ax.text3D(X[y == label, 0].mean(),
              X[y == label, 1].mean() + 1.5,
              X[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral,
           edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
```

Principal Component Analysis (PCA) applied to this data identifies the combination of attributes (principal components, or directions in the feature space) that account for the most variance in the data. Here we plot the different samples on the 2 first principal components.

Linear Discriminant Analysis (LDA) tries to identify attributes that account for the most variance between classes. In particular, LDA, in contrast to PCA, is a supervised method, using known class labels.

```python
print(__doc__)

import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

iris = datasets.load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)

# Percentage of variance explained for each components
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))

plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')

plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], alpha=.8, color=color,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of IRIS dataset')

plt.show()
```
